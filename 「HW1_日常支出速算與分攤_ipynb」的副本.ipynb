```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)
```

```python
import pandas as pd
# read data and put it in a dataframe
# 在 google 工作表載入 gsheets
gsheets = gc.open_by_url('https://docs.google.com/spreadsheets/d/1PXxNLAHsfL1_WmXYe4Jz11kk_S-thY7DzZCPNWcRA30/edit?gid=0#gid=0')
```

```python
# 從 gsheets 的 All-whiteboard-device 載入 sheets
sheets = gsheets.worksheet('工作表1').get_all_values()
# 將 sheets1 資料載入 pd 的 DataFrame 進行分析
df = pd.DataFrame(sheets[1:], columns=sheets[0])
# 取得最前面的5筆資料
df.head()
```

```python
import datetime
```

```python
# 讓使用者輸入資料
date_str = input("請輸入日期 (格式：YYYY-MM-DD): ")
# 檢查日期格式是否正確
datetime.datetime.strptime(date_str, '%Y-%m-%d')

time_str = input("請輸入時間 (格式：HH:MM): ")
# 檢查時間格式是否正確
datetime.datetime.strptime(time_str, '%H:%M')

name = input("請輸入付款人: ")
place = input("請輸入地點: ")
way = input("請輸入付款方式: ")
note = input("備註: ")
```

```python
# 讓使用者輸入品項
item = input("請輸入品項: ")

# 讓使用者輸入金額，並確保是數字
while True:
    try:
        amount = float(input("請輸入金額: "))
        break
    except ValueError:
        print("金額必須是數字，請重新輸入。")

# 讓使用者輸入分類
category_str = input("請輸入分類: ")

# 創建一個新的 DataFrame 來代表你要新增的資料
new_data = pd.DataFrame([{
    '日期': date_str,
    '時間': time_str,
    '分類': category_str, # 使用使用者輸入的分類
    '品項': item,
    '金額': amount,
    '付款人': name,       # 使用使用者輸入的付款人
    '地點': place,      # 使用使用者輸入的地點
    '支付方式': way,    # 使用使用者輸入的支付方式
    '備註': note        # 使用使用者輸入的備註
}])

# 使用 concat() 將新資料合併到舊的 df 中
# 檢查 df 是否為空，如果是，則直接賦值；否則進行合併
if df.empty:
    df = new_data
else:
    df = pd.concat([df, new_data], ignore_index=True)
```

```python
df
```

```python
# 將 DataFrame 轉換成列表的列表 (list of lists)，這是 gspread 支援的資料格式
data_to_write = df.values.tolist()

# 第一步：取得工作表物件
worksheet = gsheets.worksheet('工作表1')

# 第二步：將資料寫入工作表
worksheet.append_rows(values=data_to_write, value_input_option='USER_ENTERED')
print("資料已成功寫入 Google 工作表！")
```
